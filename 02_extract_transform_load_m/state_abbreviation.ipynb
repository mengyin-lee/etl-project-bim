```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
url = 'https://www.ssa.gov/international/coc-docs/states.html'
```

```python
tables = pd.read_html(url)
tables
```

```python
state_abbr_df = tables[0]
state_abbr_df
```

```python
rds_connection_string = "postgres:rootroot@localhost:5432/ETLproject_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
state_abbr_df.columns = ['state_name', 'state_code']
```

```python
state_abbr_df.to_sql(name='state_abbr', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from state_abbr', con=engine).head()
```
